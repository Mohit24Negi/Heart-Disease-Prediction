* The dataset was split into **training and testing sets**, with **70%** of the data used for **model training** and **30%** reserved for **model evaluation**, ensuring reproducibility through a fixed **random_state**.


### 04. Modeling

```python
# result table
result_df = pd.DataFrame(
    columns=["Model_name", "ROC_score", "Recall"]
)

def model_building(index, model, threshold=0.5):
    
    # Train model
    model.fit(x_train, y_train)
    
    # Probabilities & predictions
    y_prob = model.predict_proba(x_test)[:, 1]
    y_pred = (y_prob >= threshold).astype(int)
    
    # Metrics
    auc = roc_auc_score(y_test, y_prob)
    recall = recall_score(y_test, y_pred)
    
    # Store main results
    result_df.loc[index] = [
        model.__class__.__name__,
        auc,
        recall
    ]
    
    # Store extra outputs (optional use later)
    outputs = {
        "model": model,
        "y_prob": y_prob,
        "y_pred": y_pred,
        "confusion_matrix": confusion_matrix(y_test, y_pred),
        "classification_report": classification_report(
            y_test, y_pred, output_dict=True
        )
    }
    
    return outputs
```
